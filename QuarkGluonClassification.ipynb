```python
from mpl_toolkits import mplot3d
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d
from matplotlib import cm

quark_data = pd.read_excel("/Users/ravial-sahalman/Downloads/QuarkSamples.xlsx")
gluon_data = pd.read_excel("/Users/ravial-sahalman/Downloads/Gluons.xlsx")

column1 = []
column2 = []
column3 = []
column4 = []
for row in quark_data.itertuples():
    column1.append(row[1]) # Multiplicty
    column2.append(row[2]) # Mass
    column3.append(row[3]) # Angluarity
    column4.append(row[4]) # PTD
    
    
column5 = []
column6 = []
column7 = []
column8 = []
for row in gluon_data.itertuples():
    column5.append(row[1]) # Multiplicty
    column6.append(row[2]) # Mass
    column7.append(row[3]) # Angluarity
    column8.append(row[4]) # PTD
    
    
    
plt.hist(column1,bins = 30,alpha = 0.5,histtype = 'bar', ec = 'black', label = 'Quark')  
plt.hist(column5,bins = 30,alpha = 0.5,histtype = 'bar', ec = 'black', label = 'Gluon')
plt.xlabel("Multiplicity")
plt.legend()
plt.show()
   
plt.hist(column2,bins = 30,alpha = 0.5,histtype = 'bar', ec = 'black', label = 'Quark')
plt.hist(column6,bins = 30,alpha = 0.5,histtype = 'bar', ec = 'black', label = 'Gluon')
plt.xlabel('Mass')
plt.legend()
plt.show()

plt.hist(column3,bins = 30,alpha = 0.5,histtype = 'bar', ec = 'black', label = 'Quark')
plt.hist(column7,bins = 30,alpha = 0.5,histtype = 'bar', ec = 'black', label = 'Gluon')
plt.xlabel('Angularity')
plt.legend()
plt.show()

plt.hist(column4,bins = 30,alpha = 0.5,histtype = 'bar', ec = 'black', label = 'Quark')
plt.hist(column8,bins = 30,alpha = 0.5,histtype = 'bar', ec = 'black', label = 'Gluon')
plt.xlabel('PTD')
plt.legend()
plt.show()
    
```
